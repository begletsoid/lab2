```python
housing = pd.read_csv('housing.csv')
housing['blank'] = housing['total_bedrooms'].apply(lambda total_bedrooms: 0 if math.isnan(total_bedrooms) else 1)

DUMMY_SEP = '__'
housing = pd.get_dummies(housing, prefix_sep=DUMMY_SEP)
```

```python
housing['average_rooms'] = housing['total_rooms']/housing['households']
housing['average_bedrooms'] = housing['total_bedrooms']/housing['households']

housing = housing.drop('total_rooms', axis=1)
housing = housing.drop('total_bedrooms', axis=1)
```

```python
housing_blank = housing.loc[housing['blank'] == 0]

housing =  housing.loc[housing['blank'] == 1]
housing = housing.drop('blank', axis=1)

housing_blank = housing_blank.drop('blank', axis=1)
```

```python
housing = housing.loc[housing['average_bedrooms'] > 0.6]
housing = housing.loc[housing['average_bedrooms'] < 1.9]
```

```python
housing_train_val, housing_test = train_test_split(housing, test_size=0.2)
housing_test = housing_test.copy()
housing_test['longitude'] -= np.mean(housing_test['longitude'])
housing_test['longitude'] /= np.std(housing_test['longitude'])
housing_test['latitude'] -= np.mean(housing_test['latitude'])
housing_test['latitude'] /= np.std(housing_test['latitude'])

housing_train, housing_val = train_test_split(housing_train_val, test_size=0.5)
housing_val = housing_val.copy()
housing_val.loc[:, 'longitude'] -= np.mean(housing_val['longitude'])
housing_val.loc[:, 'longitude'] /= np.std(housing_val['longitude'])
housing_val.loc[:, 'latitude'] -= np.mean(housing_val['latitude'])
housing_val.loc[:, 'latitude'] /= np.std(housing_val['latitude'])

housing_train = housing_train.copy()
housing_train.loc[:, 'longitude'] -= np.mean(housing_train['longitude'])
housing_train.loc[:, 'longitude'] /= np.std(housing_train['longitude'])
housing_train.loc[:, 'latitude'] -= np.mean(housing_train['latitude'])
housing_train.loc[:, 'latitude'] /= np.std(housing_train['latitude'])

housing_blank = housing_blank.copy()
housing_blank.loc[:, 'longitude'] -= np.mean(housing_blank['longitude'])
housing_blank.loc[:, 'longitude'] /= np.std(housing_blank['longitude'])
housing_blank.loc[:, 'latitude'] -= np.mean(housing_blank['latitude'])
housing_blank.loc[:, 'latitude'] /= np.std(housing_blank['latitude'])
```

```python
X_train, y_train = housing_train.drop("average_bedrooms",axis=1), housing_train['average_bedrooms']

X_test, y_test = housing_test.drop("average_bedrooms",axis=1), housing_test['average_bedrooms']
X_val, y_val = housing_val.drop("average_bedrooms",axis=1), housing_val['average_bedrooms']
```

```python
linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, y_train) 

predictions = linear_regression_model.predict(X_val)
mse1 = mean_squared_error(y_train, predictions) 
print("RMSE: {}".format(np.sqrt(mse1)))
```

```python
X_blank = housing_blank.drop('average_bedrooms',axis=1)
y_blank = housing_blank['average_bedrooms']
```

```python
housing_blank['average_bedrooms'] = linear_regression_model.predict(X_blank)
```

```python
model = KMeans(n_clusters=4)
model.fit(housing_train)
housing_train['cluster'] = model.labels_
model.fit(housing_val)
housing_val['cluster'] = model.labels_

X_housing_train, y_housing_train = housing_train.drop("median_house_value",axis=1), housing_train['median_house_value']
X_housing_val, y_housing_val = housing_val.drop("median_house_value",axis=1), housing_val['median_house_value']

regressor = DecisionTreeRegressor(max_depth = 25, min_samples_split = 33) 
regressor.fit(X_housing_train, y_housing_train) 

predictions_regr = regressor.predict(X_housing_val)
mse2 = mean_squared_error(y_housing_val, predictions_regr) 
print("RMSE: {}".format(np.sqrt(mse2)))

fig, ax = plt.subplots()
ax.set_ylim(0, 600000)
ax.set_xlim(0, 600000)
ax.scatter(y_housing_val, regressor.predict(X_housing_val))
ax.plot(np.linspace(0,1000000,100), np.linspace(0,1000000,100), color='orange')
pass
```
